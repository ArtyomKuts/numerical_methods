```python
import numpy as np 
import matplotlib.pyplot as plt
import random
```

Задаём число обусловленности, размерность матрицы и решение СЛАУ

```python
cond=float(input('Задайте число обусловленности: '))
n=int(input('Задайте размерность матрицы: '))
result_X=np.random.sample(n).reshape(-1, 1)
result_X
```

Задаём элементы на главной диагонали, создаём треугольную матрицу

```python
lambda1=1
lambda2=cond
lambda_n=random.uniform(1,cond)
u=np.full(n-2,lambda_n)
v=np.concatenate((np.array([lambda1,cond]),u))
D=np.diag(v)
T=np.tri(n)
G=np.diag(np.ones(n))
B=T+D-G
```

Cоздаём ортогональную матрицу

```python
E=np.eye(n)
w=np.random.sample(n)
Q=E-2*w*w.reshape(-1, 1)/np.linalg.norm(w)**2
```

Создаём матрицу и вычисяем правую часть СЛАУ по известной матрице

```python
A=np.dot(np.linalg.inv(Q),B)
A=np.dot(A, Q)
b=np.dot(A,result_X)
A,b
```

Зададим возмущение матрицы коэффициентов

```python
F=np.zeros((n,n))
for i in range(n):
    for j in range(n):
        k=2*(0.5-random.random())*A[i][j]*0.02
        A[i][j]+=k
        F[i][j]+=k       
A,F
```

Сделаем QR разложение матрицы А

```python
Z=np.zeros((n,n))
def QR_decomposition(C,B):
    R=np.zeros((n,n))
    B[0:,0]=C[0:,0]
    B[0:,0]=B[0:,0]/np.linalg.norm(B[0:,0])
    for i in range(1,n):
        B[0:,i]=C[0:,i]
        for j in range(i):
            B[0:,i]-=np.dot(np.dot(C[0:,i],B[0:,j])/np.dot(B[0:,j],B[0:,j]),B[0:,j])
        B[0:,i]=B[0:,i]/np.linalg.norm(B[0:,i])
    for i in range(n):
        for j in range(i,n):
            R[i,j]+=np.dot(B[0:,i],C[0:,j])
    composite_matrix=np.hstack([R,E])
    j=n-1
    while j>=0:
        composite_matrix[j,j:]=composite_matrix[j,j:]/composite_matrix[j,j]
        i=j-1
        while i>=0:
            composite_matrix[i,j:]=composite_matrix[i,j:]-composite_matrix[j,j:]*composite_matrix[i,j]
            i-=1
        j-=1
    inv_R=composite_matrix[0:,n:]
    return [inv_R,B]
```

```python
X=np.dot(QR_decomposition(A,Z)[0],np.transpose(QR_decomposition(A,Z)[1]))
X=np.dot(X,b)
X
```

Вычислим норму ошибки

Вычислим норму невязки

```python
cond_A=np.linalg.norm(A,2)*np.linalg.norm(np.linalg.inv(A),2)
cond_A
```
